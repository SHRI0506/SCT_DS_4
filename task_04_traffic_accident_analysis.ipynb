```python
import pandas as pd

data = {
    "Hour": [1, 3, 8, 9, 18, 20, 22, 7, 14, 16, 19, 21, 23, 6, 10],
    "Weather": ["Clear", "Rain", "Clear", "Fog", "Clear", "Rain", "Clear",
                "Clear", "Rain", "Clear", "Fog", "Clear", "Rain", "Clear", "Fog"],
    "Road_Condition": ["Dry", "Wet", "Dry", "Wet", "Dry", "Wet", "Dry",
                       "Dry", "Wet", "Dry", "Wet", "Dry", "Wet", "Dry", "Wet"],
    "Accidents": [2, 4, 6, 5, 8, 7, 3, 5, 6, 7, 9, 4, 5, 3, 6]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.lineplot(x='Hour', y='Accidents', data=df, marker='o')
plt.title("Traffic Accidents by Time of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x='Weather', y='Accidents', data=df, estimator=sum)
plt.title("Total Accidents by Weather Condition")
plt.xlabel("Weather Condition")
plt.ylabel("Total Number of Accidents")
plt.show()

```

```python
plt.figure(figsize=(7,5))
sns.barplot(x='Road_Condition', y='Accidents', data=df, estimator=sum)
plt.title("Total Accidents by Road Condition")
plt.xlabel("Road Condition")
plt.ylabel("Total Number of Accidents")
plt.show()

```

### Conclusions

- Traffic accidents are more frequent during peak morning and evening hours.
- Poor weather conditions such as rain and fog contribute to higher accident rates.
- Wet road conditions significantly increase the risk of accidents.
- Time, weather, and road conditions play a crucial role in traffic safety.
